```python
# 데이터 처리 및 관리 솔루션 개발 공급업 
file2='../손예림/데이터처리및관리솔루션개발·공급업영역별시장규모2.csv'
```

```python
DF2=pd.read_csv(file2,encoding='utf-8')
```

```python
DF2.rename(columns={'2022 e)':'2022'},inplace=True)
```

```python
DF2[['2015','2016']] =DF2[['2015','2016']].apply(lambda x : pd.to_numeric(x,errors='coerce'))
```

```python
for row in DF2.index[20:24]:
    DF2.loc[row] = DF2.loc[row].fillna(DF2.loc[row][2:].mean())

DF2
```

```python
DF2=DF2.iloc[4::4].reset_index(drop=True)
plt.figure(figsize=(14,10))
for i in range(6) :
    plt.plot(range(8), DF2.iloc[i,2:],label=DF2.iloc[i,0],marker='o',markersize=5,linestyle='--')
plt.legend()
plt.title("공급엽영역별 시장규모")
plt.xticks(range(8), [i for i in range(2015,2023)])
plt.yscale('log',base=2)
plt.show()
```
